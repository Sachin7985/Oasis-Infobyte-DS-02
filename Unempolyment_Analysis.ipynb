Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly .express as px
```

```python
data = pd.read_csv('/content/Unemployment.csv')
```

```python
data.head()
```

```python

# checking dataset information
data.info()
```

```python
# describing the dataset
data.describe()
```

```python
# check null/missing values
data.isnull().sum()
```

```python
data.head()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8,6))

# Compute the correlation matrix and plot the heatmap
ax = sns.heatmap(data.corr(), annot=True,linewidth=3)

plt.show()
```

```python
# plotting histplot

data.columns=['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region.1','Longitude','Latitude']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region.1',data=data)
plt.show()
```

```python
# plotting histplot

plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue='Region.1',data=data)
plt.show()
```

```python

# plotting sunburst

unemployment = data[['States','Region.1','Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region.1','States'],
                     values='Estimated Unemployment Rate',
                     width=700,height=600, color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```
